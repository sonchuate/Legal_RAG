```python
import re

def merge_line(text:str):
    lines = text.split('\n')
    lines = [line.strip() for line in lines if len(line.strip()) > 0]

    result = []
    tmp = []
    len_lines = len(lines)
    for id, line in enumerate(lines):
        tmp.append(line)
        if line.strip()[-1] in ";.!?:":
            result.append(" ".join(tmp))
            tmp = []
        elif id + 1 < len_lines:
            if lines[id + 1][0].isupper() or lines[id + 1][0].isnumeric():
                result.append(" ".join(tmp))
                tmp = []
    return "\n".join(result)

with open('thu_vien_phap_luat/meta.txt', 'r', encoding='utf-8') as f:
    data = f.read()
for line in data.split('\n'):
    try:
        if len(line) < 3:
            continue
        link, id = line.split('\t')
        r = rq.get(link)
        soup = BeautifulSoup(r.text)
        with open(f'thu_vien_phap_luat/{id}.txt', 'w', encoding='utf-8') as f: 
        # print(str(soup.find('div', id='tab1')))
            for doc in str(soup.find('div', id='tab1')).split('name="dieu_'):
                f.write(f"""[[Điều {doc.split('"')[0]}]]\n""")
                f.write(merge_line(">".join(re.sub(r"</?\w+[^>]*>", "", doc).split('>')[1:]).split("<")[0]))
                f.write(f'\n')
    except:
        print(line)
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import json
```

```python
for id_page in range(2, 100, 1):
    page  = f"https://www.topcv.vn/viec-lam-it?page={id_page}"
    driver = webdriver.Chrome()
    driver.get(page)
    elements = driver.find_elements(By.XPATH, '//*[@data-box="BoxSearchResult"]')

    for element in elements:
        # print(element.text)
        try:
            # Tìm thẻ <a> có class="company" bên trong element (dù không là con trực tiếp)
            company_link = element.find_element(By.XPATH, './/a[contains(@class, "company")]')
            
            # Lấy URL của thẻ <a>
            href = company_link.get_attribute("href")
            company_name = company_link.text

            title = element.find_element(By.XPATH, './/h3[contains(@class, "title")]')
            link = title.find_element(By.XPATH, './/a[contains(@target, "_blank")]').get_attribute("href")
            title = title.text

            deadline = element.find_element(By.XPATH, './/label[contains(@class, "deadline")]').text

            address = element.find_element(By.XPATH, './/label[contains(@class, "address")]').text

            time_apply = element.find_element(By.XPATH, './/label[contains(@class, "time")]').text

            labels = element.find_elements(By.XPATH, './/label[contains(@class, "item")]')

            labels = [l.text for l in labels]
            data = {
                "href": href,
                "company_name": company_name,
                "title": title,
                "deadline": deadline,
                "address": address,
                "time_apply": time_apply,
                "labels": labels,
                "link": link
            }

            with open(f"E:/datn/data/{company_name}_{title}.json", "w", encoding="utf-8") as f:
                json.dump(data, f, ensure_ascii=False, indent=4)

        except:
            pass
        
        # try:
        #     # Tìm thẻ <a> có class="company" bên trong element (dù không là con trực tiếp)
        #     company_link = element.find_element(By.XPATH, './/a[contains(@class, "company")]')
            
        #     # Lấy URL của thẻ <a>
        #     href = company_link.get_attribute("href")
        #     print(href)

        # except:
        #     print("Không tìm thấy thẻ <a class='company'> trong phần tử này.")
    driver.quit()
```

```python
for file in os.listdir('E:/datn/data'):
    if not file.endswith('.json'): continue
    

    try:
        input_path = f"E:/datn/data/{file}"
        with open(input_path, "r", encoding="utf-8") as f:
            data = json.load(f)
        page = data['link']
        company_name = data['company_name']
        title = data['title']
        if os.path.exists(f"E:/datn/data1/{company_name}_{title}.txt"): continue
        driver = webdriver.Chrome()
        driver.get(page)
        jd = driver.find_element(By.CLASS_NAME, "job-detail__information-detail--content").text.strip()
        with open(f"E:/datn/data1/{company_name}_{title}.txt", "w", encoding="utf-8") as f:
            f.write(jd)
        driver.quit()
    except:
        print(file)
```
