```python
from pymilvus import MilvusClient

client = MilvusClient("milvus_demo.db")
```

```python
if client.has_collection(collection_name="demo_collection"):
    client.drop_collection(collection_name="demo_collection")
client.create_collection(
    collection_name="demo_collection",
    dimension=1024,  # The vectors we will use in this demo has 768 dimensions
)
```

```python
import random

# Text strings to search from.
docs = [
    "Artificial intelligence was founded as an academic discipline in 1956.",
    "Alan Turing was the first person to conduct substantial research in AI.",
    "Born in Maida Vale, London, Turing was raised in southern England.",
]
# Use fake representation with random vectors (768 dimension).
vectors = [[random.uniform(-1, 1) for _ in range(1024)] for _ in docs]
data = [
    {"id": i, "vector": vectors[i], "text": docs[i]}
    for i in range(len(vectors))
]

print("Data has", len(data), "entities, each with fields: ", data[0].keys())
print("Vector dim:", len(data[0]["vector"]))
```

```python
res = client.insert(collection_name="demo_collection", data=data)

print(res)
```

```python
from FlagEmbedding import BGEM3FlagModel

model = BGEM3FlagModel('BAAI/bge-m3',
                       use_fp16=True) # Setting use_fp16 to True speeds up computation with a slight performance degradation

sentences_1 = ["What is BGE M3?", "Defination of BM25"]
sentences_2 = ["BGE M3 is an embedding model supporting dense retrieval, lexical matching and multi-vector interaction.",
               "BM25 is a bag-of-words retrieval function that ranks a set of documents based on the query terms appearing in each document"]

embeddings_1 = model.encode(sentences_1,
                            batch_size=12,
                            max_length=8192, # If you don't need such a long length, you can set a smaller value to speed up the encoding process.
                            )['dense_vecs']
embeddings_2 = model.encode(sentences_2)['dense_vecs']
similarity = embeddings_1 @ embeddings_2.T
print(similarity)
# [[0.6265, 0.3477], [0.3499, 0.678 ]]
```

```python
def embed(sentences, batch_size=256):
    return model.encode(sentences,
                            batch_size=batch_size,
                            max_length=8192, # If you don't need such a long length, you can set a smaller value to speed up the encoding process.
                            )['dense_vecs']
print(embed(["Cô ấy là một người vui_tính .", "Cô ấy cười nói suốt cả ngày ."]))
```

# Indexing

```python
if client.has_collection(collection_name="demo_collection"):
    client.drop_collection(collection_name="demo_collection")
client.create_collection(
    collection_name="demo_collection",
    dimension=1024,  # The vectors we will use in this demo has 768 dimensions
)
```

```python
import re
from tqdm import tqdm
import numpy as np
def replace_brackets(text):
    return re.sub(r"\[\[.*?\]\]", "|||", text) # thay thế [[ abc ]] thành |||

import os

BATCH_SIZE = 16

PATH_TO_DATA = '/kaggle/input/legal-crawl-v1/thu_vien_phap_luat'

docs = []
id = 0
for file in tqdm(os.listdir(PATH_TO_DATA)):
    with open(os.path.join(PATH_TO_DATA, file), 'r', encoding='utf-8') as f:
        data = f.read()
        data = replace_brackets(data)
        data = [d.strip() for d in data.split('|||') if len(d) > 3]
        docs.extend(data)
    if len(docs) >= BATCH_SIZE:
        embed_vectors = embed(docs)
        data = []
        for e, d in zip(embed_vectors, docs):
            data.append({"id" : id, "vector": np.array(e, dtype=np.float32), "text": d})
            id += 1
        client.insert(collection_name="demo_collection", data=data)
        docs = []
```

```python
query = "độ tuổi được sử dụng xe gắn máy"
res = client.search(
    collection_name="demo_collection",  # target collection
    data=np.array(embed([query]), dtype=np.float32),  # query vectors
    limit=20,  # number of returned entities
    output_fields=["text"],  # specifies fields to be returned
)

print(res)
```
